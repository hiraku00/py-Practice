<p><b>matplotlib.pyplot.Line2D クラスの主要な引数</b></p>
<table>
<tr>
<th>xdata (必須)</th>
<td>X 軸方向の数値</td>
</tr>
<tr>
<th>ydata (必須)</th>
<td>Y 軸方向の数値</td>
</tr>
<tr>
<th>linewidth</th>
<td>線の太さ</td>
</tr>
<tr>
<th>linestyle</th>
<td>線のスタイル。‘solid’ (実線), ‘dashed’ (破線), ‘dashdot’ (破線&#038;点線), ‘dotted’ (点線) から指定。(デフォルト値:&#8221;solid&#8221;) </td>
</tr>
<tr>
<th>color</th>
<td>線の色</td>
</tr>
<tr>
<th>marker</th>
<td>マーカーの種類。参考: <a href="http://matplotlib.org/api/markers_api.html#module-matplotlib.markers">matplotlib.markers</a> (デフォルト値:&#8221;None&#8221;)</td>
</tr>
<tr>
<th>markersize</th>
<td>マーカーの大きさ</td>
</tr>
<tr>
<th>markeredgewidth</th>
<td>マーカーの枠線の太さ</td>
</tr>
<tr>
<th>markeredgecolor</th>
<td>マーカーの枠線の色</td>
</tr>
<tr>
<th>markerfacecolor</th>
<td>マーカーの塗りつぶしの色</td>
</tr>
<tr>
<th>markerfacecoloralt</th>
<td>マーカーの塗りつぶしの色 2。fillstyle で left, right, bottom, top を指定した際、塗りつぶされない領域が &#8216;markerfacecoloralt&#8217; で指定された色となります。 (デフォルト値: &#8216;none&#8217;)</td>
</tr>
<tr>
<th>fillstyle</th>
<td>マーカーの塗りつぶしのスタイル。‘full’ (全体), ‘left’ (左半分), ‘right’ (右半分), ‘bottom’ (下半分), ‘top’ (上半分), ‘none’ (塗りつぶしなし) から選択。</td>
</tr>
<tr>
<th>antialiased</th>
<td>アンチエイリアス (線を滑らかに描画する処理) を適用するかどうか。False または True から選択。</td>
</tr>
</table>
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

# 折れ線グラフを出力
left = np.array([1,2,3,4,5])
height = np.array([100,300,200,500,400])
plt.plot(left, height)
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
線の太さや色など見た目の設定
</div>
<div style="font-size:14px; color:white; margin:20px">
線の太さを 4、線の色を赤 (red) に指定
</div>

```python
plt.plot(left, height, linewidth=4, color='red')
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
複数の線を出力。上から、実線、破線、破線&点線、点線。
</div>

```python
plt.plot(left, height, linestyle="solid")
plt.plot(left, height/2, linestyle="dashed")
plt.plot(left, height/3, linestyle="dashdot")
plt.plot(left, height/4, linestyle="dotted")
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
マーカーを表示
</div>
<div style="font-size:14px; color:white; margin:20px">
マーカー（丸印: o）を出力
</div>

```python
plt.plot(left, height, marker="o")
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
マーカー（丸印: o）、枠線なしで出力
</div>

```python
plt.plot(left, height, marker="o", markeredgewidth="0")
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
マーカーをダイヤモンド型 (D)、サイズ 12、マーカーの枠線のサイズ 3、<br>
マーカーの枠線の色 青、マーカーの塗りつぶしの色 水色で出力。
</div>

```python
plt.plot(left, height, marker="D", markersize=12, markeredgewidth=3, markeredgecolor="blue",
         markerfacecolor="lightblue")
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
マーカーをダイヤモンド型 (s)、サイズ 20、マーカーの枠線のサイズ 3、<br>
マーカーの枠線の色 黒、マーカーの塗りつぶしの色1 青、マーカーの塗りつぶしの色2 赤、マーカーの塗りつぶし方法 左半分で出力。
</div>

```python
plt.plot(left, height, marker="s", markersize=20, markeredgewidth=3, markeredgecolor="black",
         markerfacecolor="blue", markerfacecoloralt="red", fillstyle="left")
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
アンチエイリアス (線を滑らかにする処理) を無効化 (上側)
</div>

```python
plt.plot(left, height, linewidth=5, antialiased=False)
plt.plot(left, height/2, linewidth=5)
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
凡例、タイトル、ラベルなど
</div>
<div style="font-size:14px; color:white; margin:20px">
グラフのタイトル、X 軸、Y 軸の名前 (ラベル)、グリッド線を表示
</div>

```python
plt.plot(left, height)
plt.title("This is a title")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
左上 (loc=2) に凡例を追加
</div>

```python
p1 = plt.plot(left, height, linewidth=2)
p2 = plt.plot(left, height/2, linewidth=2, linestyle="dashed")
plt.legend((p1[0], p2[0]), ("Class 1", "Class 2"), loc=2)
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
棒グラフと折れ線グラフを混在して出力
</div>
<div style="font-size:14px; color:white; margin:20px">
</div>

```python
bar_height = np.array([100,200,300,400,500])
line_height = np.array([100,300,200,500,400])

# 棒グラフを出力
fig, ax1 = plt.subplots()
ax1.bar(left, bar_height, align="center", color="royalblue", linewidth=0)
ax1.set_ylabel('Axis for bar')

# 折れ線グラフを出力
ax2 = ax1.twinx()
ax2.plot(left, line_height, linewidth=4, color="crimson")
ax2.set_ylabel('Axis for line')

plt.show()
```
