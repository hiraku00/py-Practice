<p><b>matplotlib.pyplot.pie の主要な引数</b></p>
<table>
<tr>
<th>x (必須)</th>
<td>各要素の大きさを配列で指定。</td>
</tr>
<tr>
<th>explode</th>
<td>各要素を中心から離して目立つように表示。</td>
</tr>
<tr>
<th>labels</th>
<td>各要素のラベル。</td>
</tr>
<tr>
<th>colors</th>
<td>各要素の色を指定。</td>
</tr>
<tr>
<th>autopct</th>
<td>構成割合をパーセンテージで表示。 (デフォルト値: None)</td>
</tr>
<tr>
<th>pctdistance</th>
<td>上記のパーセンテージを出力する位置。円の中心 (0.0) から円周 (1.0) を目安に指定。autopct を指定した場合のみ有効。 (デフォルト値: 0.6)</td>
</tr>
<tr>
<th>shadow</th>
<td>True に設定すると影を表示。 (デフォルト値: False)</td>
</tr>
<tr>
<th>labeldistance</th>
<td>ラベルを表示する位置。円の中心 (0.0) から円周 (1.0) を目安に指定。 (デフォルト値: 1.1)</td>
</tr>
<tr>
<th>startangle</th>
<td>各要素の出力を開始する角度。 (デフォルト値: None)</td>
</tr>
<tr>
<th>radius</th>
<td>円の半径。 (デフォルト値: 1)</td>
</tr>
<tr>
<th>counterclock</th>
<td>True に設定すると時計回りで出力。False に設定すると反時計回りで出力。 (デフォルト値: True)</td>
</tr>
<tr>
<th>wedgeprops</th>
<td>ウェッジ (くさび形の部分) に関する指定。枠線の太さなどを設定可能。 (デフォルト値: None)</td>
</tr>
<tr>
<th>textprops</th>
<td>テキストに関するプロパティ。 (デフォルト値: None)</td>
</tr>
</table>

<div style="font-size:20px; color:pink; margin:20px">
グラフの出力例
</div>
<div style="font-size:14px; color:white; margin:20px">
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

# 円グラフを描写
x = np.array([100,200,300,400,500])
plt.pie(x)
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
スタイルに関するカスタマイズ
</div>
<div style="font-size:14px; color:white; margin:20px">
ラベルを表示
</div>

```python
label = ["Apple", "Banana", "Orange", "Grape", "Strawberry"]
plt.pie(x, labels=label)
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
開始角度を 90 度, 時計回りに設定
</div>

```python
plt.pie(x, labels=label, counterclock=False, startangle=90)
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
環境によっては必要なカスタマイズ
</div>
<div style="font-size:14px; color:white; margin:20px">
真円になるようにアスペクト比を固定<br>
plt.axis('equal') を指定することで、Jupyter等の環境で楕円のようなつぶれた形で出力されることを防ぐことができます。</div>

```python
plt.pie(x, labels=label, counterclock=False, startangle=90)
plt.axis('equal')
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
背景を白に設定
PyCharm など一部の環境では、背景がグレーで出力されるため、以下の方法で背景色を設定します。
</div>

```python
fig = plt.figure()
fig.patch.set_facecolor('white')
plt.pie(x, labels=label, counterclock=False, startangle=90)
plt.axis('equal')
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
各要素のスタイルのカスタマイズ
</div>
<div style="font-size:14px; color:white; margin:20px">
1 つめの要素を 20% ずらして目立たせて表示
</div>

```python
plt.pie(x, labels=label, counterclock=False, startangle=90,
        explode=[0.2, 0, 0, 0, 0])
plt.axis('equal')
plt.show()
```

```python
colors = ["lightpink", "yellow", "gold", "slateblue", "lightcoral"]
plt.pie(x, labels=label, counterclock=False, startangle=90, colors=colors)
plt.axis('equal')
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
各要素の色を濃淡で指定
</div>

```python
colors2 = ["0.1", "0.3", "0.5", "0.7", "0.9"]
plt.pie(x, labels=label, counterclock=False, startangle=90, colors=colors2)
plt.axis('equal')
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
影を設定
</div>

```python
plt.pie(x, labels=label, counterclock=False, startangle=90, shadow=True)
plt.axis('equal')
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
ラベルを中心から円周の間の 50% の位置に表示
</div>

```python
plt.pie(x, labels=label, counterclock=False, startangle=90, labeldistance=0.5)
plt.axis('equal')
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
各要素のスタイルに関する設定
</div>
<div style="font-size:14px; color:white; margin:20px">
枠線の太さを 3, 線の色を白に指定
</div>

```python
plt.pie(x, labels=label, counterclock=False, startangle=90,
        wedgeprops={'linewidth':3, 'edgecolor':"white"})
plt.axis('equal')
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
枠線の太さを 0 (枠線なし) に指定
</div>

```python
plt.pie(x, labels=label, counterclock=False, startangle=90,
        wedgeprops={'linewidth':0})
plt.axis('equal')
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
ラベル・テキストに関する設定
</div>
<div style="font-size:14px; color:white; margin:20px">
テキストの色を白色、太字に設定
</div>

```python
plt.pie(x, labels=label, counterclock=False, startangle=90, labeldistance=0.5,
        textprops={'color': "white", 'weight': "bold"})
plt.axis('equal')
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
構成割合をプロット。小数点以下 1 桁まで出力
</div>

```python
plt.pie(x, labels=label, counterclock=False, startangle=90, autopct="%1.1f%%")
plt.axis('equal')
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
構成割合をプロット。小数点以下 1 桁まで、中心から円周の間の 70% の位置に出力
</div>

```python
plt.pie(x, labels=label, counterclock=False, startangle=90, autopct="%1.1f%%", pctdistance=0.7)
plt.axis('equal')
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
より高度なパラメータの設定
</div>
<div style="font-size:14px; color:white; margin:20px">
ラベルの横位置を各要素の中央に合わせる
</div>

```python
patches, texts = plt.pie(x, labels=label, counterclock=False, startangle=90,
                         labeldistance=0.5,
                         textprops={'color': "white", 'weight': "bold"})
for t in texts:
    t.set_horizontalalignment('center')

plt.axis('equal')
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
ラベルの横位置を各要素の中央に合わせ、サイズを 18 に指定
</div>

```python
patches, texts = plt.pie(x, labels=label, counterclock=False, startangle=90,
                         labeldistance=0.6,
                         textprops={'color': "white", 'weight': "bold"})
for t in texts:
    t.set_horizontalalignment('center')
    t.set_size(18)

plt.axis('equal')
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
ドーナツチャート (ドーナツグラフ)
</div>
<div style="font-size:14px; color:white; margin:20px">
内側に白色で円を描画することで、ドーナツグラフを作成できます。(参考)
</div>

```python
plt.pie(x, labels=label, counterclock=False, startangle=90)

# 中心(0,0)に60%の大きさで円を描画
center_circle = plt.Circle((0,0),0.6,color='black', fc='white', linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(center_circle)
plt.axis('equal')
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
<p>2 重ドーナツチャート (2 重ドーナツグラフ)</p>
円グラフの中に円グラフを出力することで、2 重ドーナツチャートを出力できます。<br>
radius パラメータで、内側の円グラフの大きさを調整しています。<br>
</div>

```python
colors = ["blue", "orange", "green", "red", "purple"]

# 円グラフ(外側)
x1 = np.array([100,200,300,400,500])
plt.pie(x1, labels=label, counterclock=False, startangle=90, colors=colors)

# 円グラフ(内側 半径70%で描画)
x2 = np.array([150,250,300,350,450])
plt.pie(x2, counterclock=False, startangle=90, radius=0.7, colors=colors)

# 中心(0,0)に40%の大きさで円を描画
center_circle = plt.Circle((0,0),0.4,color='black', fc='white', linewidth=1.25)
fig = plt.gcf()
fig.gca().add_artist(center_circle)

plt.axis('equal')
plt.show()
```
