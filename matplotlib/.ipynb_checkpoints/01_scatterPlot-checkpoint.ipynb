<div style="font-size:20px; color:pink; margin:20px">
matplotlib でグラフ作成
</div>
<div style="font-size:14px; color:white; margin:20px">
matplotlib(読み方は、マットプロットリブ)は、Pythonのグラフ作成ライブラリで、各種グラフを作成・可視化することができます。<br>
本ページでは、matplotlib を利用して、読み込んだデータセットや集計・分析した結果を可視化する方法を紹介します。<br>|
</div>

<div style="font-size:20px; color:pink; margin:20px">
matplotlib の読み込み
</div>
<div style="font-size:14px; color:white; margin:20px">
matplotlib を使うには、以下のコードで、matplotlib.pyplot クラスを plt という名前で読み込みます。
</div>

```python
import matplotlib.pyplot as plt
```

<div style="font-size:20px; color:pink; margin:20px">
Jupyter Notebook で matplotlib のグラフを出力する
</div>
<div style="font-size:14px; color:white; margin:20px">
Pythonのノートブック形式の開発環境Jupyter Notebookを用いてmatplotlibやSeabornのグラフを出力するには、<br>
先頭に、%matplotlib inline という行を記載する必要があります。<br>
<br>
Jupyter Notebook でグラフが表示できない場合、正しく記載されているかを確認してみましょう。<br>
</div>

<div style="font-size:20px; color:pink; margin:20px">
例: Jupyter Notebook で散布図を描画
</div>
<div style="font-size:14px; color:white; margin:20px">
</div>

```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

# 乱数を生成
x = np.random.rand(100)
y = np.random.rand(100)

# 散布図を描画
plt.scatter(x, y)
plt.show()
```

<p><b>matplotlib.pyplot.scatter の主要な引数</b></p>
<table>
<tr>
<th>x, y</th>
<td>グラフに出力するデータ</td>
</tr>

<td>サイズ (デフォルト値: 20)</td>

<td>色、または、連続した色の値</td>
</tr>

<th>marker</th>
<td>マーカーの形 (デフォルト値: &#8216;o&#8217;= 円)</td>

<tr>
<th>cmap</th>
<td>カラーマップ。c が float 型の場合のみ利用可能です。</td>
</tr>
<tr>
<th>norm</th>

<tr>

<td>正規化時の最大、最小値。 指定しない場合、データの最大・最小値となります。norm にインスタンスを指定した場合、vmin, vmax の指定は無視されます。</td>
</tr>
<tr>
<th>alpha</th>
<td>透明度。0(透明)～1(不透明)の間の数値を指定。</td>
</tr>
<tr>
<th>linewidths</th>

</tr>

<div style="font-size:20px; color:pink; margin:20px">
グラフの出力例
</div>
<div style="font-size:14px; color:white; margin:20px">
以下例では、100 個 × 2 軸の乱数を2次元座標上にプロットします。
</div>

```python
import numpy as np
import matplotlib.pyplot as plt

# 乱数を生成
x = np.random.rand(100)
y = np.random.rand(100)

# 散布図を描画
plt.scatter(x, y)
plt.show()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
サイズ、色、不透明度、線のサイズ、色を指定
</div>

```python
plt.scatter(x, y, s=200, c='pink', alpha=0.5, linewidths="2", edgecolors="red")
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
マーカーを指定
</div>

```python
plt.scatter(x, y, s=200, c='yellow', marker="*", alpha=0.5, linewidths="2", edgecolors="orange")
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
グラフのタイトル、X軸、Y軸の名前(ラベル)、グリッド線を表示
</div>

```python
plt.scatter(x, y)
plt.title("This is a title")
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.grid(True)
```

<div style="font-size:20px; color:pink; margin:20px">
カラーマップを指定して、値に応じてマーカーを着色
</div>
<div style="font-size:14px; color:white; margin:20px">
引数 s の値の大小に応じて、色の濃淡やグラデーションで表現することができます。
</div>

```python
# 乱数を100件生成
value = np.random.rand(100)

# 散布図を表示
plt.scatter(x, y, s=100, c=value, cmap='Blues')

# カラーバーを表示
plt.colorbar()
```

<div style="font-size:20px; color:pink; margin:20px">
</div>
<div style="font-size:14px; color:white; margin:20px">
上記に加えて正規化における最大値 (0.6)、最小値 (0.4) を指定<br>
(右側の凡例の目盛が変わっているのがわかるかと思います)<br>
</div>

```python
plt.scatter(x, y, s=100, c=value, cmap='Blues', vmin=0.4, vmax=0.6)
plt.colorbar()
```
