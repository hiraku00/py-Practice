<div style="font-size:20px; color:pink; margin:20px">
Iris (アヤメの計測データ、通称：アイリス)
</div>
<div style="font-size:14px; color:white; margin:20px">
“setosa”, “versicolor”, “virginica” という<br>
3 種類の品種のアヤメのがく片 (Sepal)、花弁 (Petal) の幅および長さを計測したデータです。<br>
<br>
<p><strong>データセットの詳細</strong></p>
<table>
<tr>
<th>レコード数</th>
<td>150</td>
</tr>
<tr>
<th>カラム数</th>
<td>4</td>
</tr>
<tr>
<th>主な用途</th>
<td>分類 (Classification)</td>
</tr>
<tr>
<th>データセットの詳細</th>
<td><a href="https://archive.ics.uci.edu/ml/datasets/Iris">UCI Machine Learning Repository: Iris Data Set</a></td>
</tr>
</table>
</table>
<p><strong>各カラムの構成</strong></p>
<table>
<tr>
<th>sepal length (cm)</th>
<td>がく片の長さ</td>
</tr>
<tr>
<th>sepal width (cm)</th>
<td>がく片の幅</td>
</tr>
<tr>
<th>petal length (cm)</th>
<td>花弁の長さ</td>
</tr>
<tr>
<th>petal width (cm)</th>
<td>花弁の幅</td>
</tr>
</table>
</div>

```python
# データセット読み込み
from sklearn.datasets import load_iris
iris = load_iris()

# Pandasのデータフレームとして表示
import pandas as pd
pd.DataFrame(iris.data, columns=iris.feature_names)
```

```python
# 各データの分類(0: 'setosa', 1: 'versicolor', 2: 'virginica')
iris.target
```

```python
iris.target_names
```
