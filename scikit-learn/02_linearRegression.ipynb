線形回帰とは
</div>
<div style="font-size:14px; color:white; margin:20px">
線形回帰モデル (Linear Regression) とは、以下のような回帰式を用いて、<br>
説明変数の値から目的変数の値を予測するモデルです。

<p><a href="http://pythondatascience.plavox.info/wp-content/uploads/2016/07/linearregression3.png" rel="attachment wp-att-1176"><img src="http://pythondatascience.plavox.info/wp-content/uploads/2016/07/linearregression3-1024x232.png" alt="linearregression3" width="640" height="145" class="alignnone size-large wp-image-1176" srcset="https://pythondatascience.plavox.info/wp-content/uploads/2016/07/linearregression3-300x68.png 300w, https://pythondatascience.plavox.info/wp-content/uploads/2016/07/linearregression3-768x174.png 768w, https://pythondatascience.plavox.info/wp-content/uploads/2016/07/linearregression3-1024x232.png 1024w, https://pythondatascience.plavox.info/wp-content/uploads/2016/07/linearregression3.png 1164w" sizes="(max-width: 640px) 100vw, 640px" /></a></p>
<br>
特に、説明変数が 1 つだけの場合「単回帰分析」と呼ばれ、説明変数が 2 変数以上で構成される場合「重回帰分析」と呼ばれます。
</div>

<div style="font-size:14px; color:white; margin:20px">
データセットの読み込み<br>
<br>
上記で説明したデータセット (winequality-red.csv) をダウンロードし、<br>
プログラムと同じフォルダに配置後、以下コードを実行し Pandas のデータフレームとして読み込みます。<br>
</div>

```python
import pandas as pd
import numpy as np

wine = pd.read_csv("winequality-red.csv", sep=";")
wine.head()
```

<div style="font-size:20px; color:pink; margin:20px">
scikit-learn を用いた線形回帰の実行例: 単回帰分析
</div>
<div style="font-size:14px; color:white; margin:20px">
まずは、1 つの説明変数を用いて単回帰分析を行います。
</div>

```python
# sklearn.linear_model.LinearRegression クラスを読み込み
from sklearn import linear_model
clf = linear_model.LinearRegression()

# 説明変数に "density (濃度)" を利用
X = wine.loc[:, ['density']].as_matrix()

# 目的変数に "alcohol (アルコール度数)" を利用
Y = wine['alcohol'].as_matrix()

# 予測モデルを作成
clf.fit(X, Y)

# 回帰係数
print(clf.coef_)

# 切片(誤差)
print(clf.intercept_)

# 決定係数
print(clf.score(X, Y))
```

<div style="font-size:14px; color:white; margin:20px">
結果より、以下のように回帰式が求められたことがわかります。<br>
[alcohol] = -280.16382307 × [density] + 289.675343383<br>
結果を 2 次元座標上にプロットすると、以下のようになります。青線が回帰直線を表します。<br>
</div>

```python
%matplotlib inline
# matplotlib パッケージを読み込み
import matplotlib.pyplot as plt
 
# 散布図
plt.scatter(X, Y)
 
# 回帰直線
plt.plot(X, clf.predict(X))
```

<div style="font-size:20px; color:pink; margin:20px">
scikit-learn を用いた線形回帰の実行例: 重回帰分析
</div>
<div style="font-size:14px; color:white; margin:20px">
続いて、「quality」を目的変数に、「quality」以外を説明変数として、重回帰分析を行います。
</div>

```python
from sklearn import linear_model
clf = linear_model.LinearRegression()

# 説明変数に"quality(品質スコア以外全て)"を利用
wine_except_quality = wine.drop("quality", axis=1)
X = wine_except_quality.as_matrix()

# 目的変数に"quality(品質スコア)"を利用
Y = wine['quality'].as_matrix()

# 予測モデルを作成
clf.fit(X, Y)

# 偏回帰係数
print(pd.DataFrame({"Name":wine_except_quality.columns,
                    "Coefficients":clf.coef_}).sort_values(by='Coefficients'))

# 切片(誤差)
print(clf.intercept_)
```

<div style="font-size:20px; color:pink; margin:20px">
scikit-learn を用いた線形回帰の実行例: 各変数を正規化して重回帰分析
</div>
<div style="font-size:14px; color:white; margin:20px">
各変数がどの程度目的変数に影響しているかを確認するには、<br>
各変数を正規化 (標準化) し、平均 = 0, 標準偏差 = 1 になるように変換した上で、<br>
重回帰分析を行うと偏回帰係数の大小で比較することができるようになります。
</div>

```python
# データフレームの各列を正規化
wine2 = wine.apply(lambda x: (x-np.mean(x)) / (np.max(x) - np.min(x)))
wine2.head()
```

```python
# 説明変数に "quality (品質スコア以外すべて)" を利用
wine2_except_quality = wine2.drop("quality", axis=1)
X = wine2_except_quality.as_matrix()

# 目的変数に "quality (品質スコア)" を利用
Y = wine2['quality'].as_matrix()

# 予測モデルを作成
clf.fit(X, Y)

# 偏回帰係数
print(pd.DataFrame({"Name":wine2_except_quality.columns,
                    "Coefficients":np.abs(clf.coef_)}).sort_values(by='Coefficients'))

# 切片
print(clf.intercept_)
```
