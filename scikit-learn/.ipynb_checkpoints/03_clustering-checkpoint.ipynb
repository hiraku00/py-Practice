<div style="font-size:20px; color:pink; margin:20px">
クラスタ分析とは
</div>
<div style="font-size:14px; color:white; margin:20px">
<p><strong>クラスタ分析 (クラスタリング, Clustering) </strong>とは、ラベル付けがなされていないデータに対して、近しい属性を持つデータをグループ化する手法です。例をあげると、以下のような活用方法があり、マーケティング施策や商品の企画開発などに活用することます。</p>
<ul>
<li>製品ごとの特徴 (自動車であれば、価格や定員、燃費、排気量、直近の販売台数) を用いて類似の製品をグループ化</li>
<li>店舗の特徴 (スーパーであれば、売上や面積、従業員数、来客数、駐車場の数) から類似の店舗をグループ化</li>
<li>顧客の特徴 (銀行であれば、性別、年齢、貯蓄残高、毎月の支出、住宅ローンの利用有無など) を用いて似たような利用傾向の顧客をグループ化</li>
</ul>
<p>クラスタ分析には大別して、K-Means に代表される「<strong>非階層的クラスタ分析</strong>」と Ward 法 (ウォード法) に代表される「<strong>階層的クラスタリング</strong>」の2種類が存在します。本ページでは、非階層的クラスタ分析の代表例である K-Means 法を用いたクラスタリングについて解説します。</p>
<p><strong>非階層的クラスタリング</strong><br />
非階層的クラスタリング (例: K-Means 法) では、決められたクラスタ数にしたがって、近い属性のデータをグループ化します。以下の図では、3つのクラスタに分類しましたが、それぞれの色でどのクラスタに分類されたかを示しています。<br />
<a href="http://pythondatascience.plavox.info/wp-content/uploads/2016/05/kmeans.png" rel="attachment wp-att-407"><img src="http://pythondatascience.plavox.info/wp-content/uploads/2016/05/kmeans-300x235.png" alt="kmeans" width="300" height="235" class="alignnone size-medium wp-image-407" srcset="https://pythondatascience.plavox.info/wp-content/uploads/2016/05/kmeans-300x235.png 300w, https://pythondatascience.plavox.info/wp-content/uploads/2016/05/kmeans.png 532w" sizes="(max-width: 300px) 100vw, 300px" /></a></p>
<p><strong>階層的クラスタリング</strong><br />
階層的クラスタリング (例: Ward 法) では、クラスタリングの結果を木構造で出力する特徴があります。縦方向の長さ (深さ) は類似度を示し、長いほど類似度が低く、短いほど類似度が高いことを示します。<br />
<a href="http://pythondatascience.plavox.info/wp-content/uploads/2016/05/ward.png" rel="attachment wp-att-408"><img src="http://pythondatascience.plavox.info/wp-content/uploads/2016/05/ward-300x244.png" alt="ward" width="300" height="244" class="alignnone size-medium wp-image-408" srcset="https://pythondatascience.plavox.info/wp-content/uploads/2016/05/ward-300x244.png 300w, https://pythondatascience.plavox.info/wp-content/uploads/2016/05/ward.png 625w" sizes="(max-width: 300px) 100vw, 300px" /></a></p>
<hr>
</div>

<div style="font-size:20px; color:pink; margin:20px">
K-Means法とは
</div>
<div style="font-size:14px; color:white; margin:20px">
K-Means 法 (K-平均法ともいいます) は、基本的には、以下の 3 つの手順でクラスタリングを行います。<br>
<br>
初期値となる重心点をサンプルデータ (データセット全体からランダムに集めた少量のデータ) から決定。<br>
各サンプルから最も近い距離にある重心点を計算によって求め、クラスタを構成。<br>
2.で求めたクラスタごとに重心を求め、2. を再度実行する。2. ～ 3. を決められた回数繰り返し実行し、<br>
大きな変化がなくなるまで計算。
</div>

<p></p>
<p><b>sklearn.cluster.KMeans クラスの引数</b><br />
実行時に、以下のパラメータを制御できます。</p>
<table>
<tr>
<th>n_clusters</th>
<td>クラスタ数。(デフォルト値: 8)</td>
</tr>
<tr>
<th>max_iter</th>
<td>繰り返し回数の最大値。 (デフォルト値: 300)</td>
</tr>
<tr>
<th>n_init</th>
<td>初期値選択において、異なる乱数のシードで初期の重心を選ぶ処理の実行回数。 (デフォルト値: 10)</td>
</tr>
<tr>
<th>init</th>
<td>初期化の方法。&#8217;k-means++&#8221;, &#8216;random&#8217; もしくは ndarray を指定。 (デフォルト値: &#8216;k-means++&#8217;)</td>
</tr>
<tr>
<th>tol</th>
<td>収束判定に用いる許容可能誤差。 (デフォルト値: 0.0001)</td>
</tr>
<tr>
<th>precompute_distances</th>
<td>距離 (データのばらつき具合) を事前に計算するか。 &#8216;auto&#8217;, True, False から指定。 (デフォルト値: &#8216;auto&#8217;)</td>
</tr>
<tr>
<th>verbose</th>
<td>1 を指定すると、詳細な分析結果を表示。 (デフォルト値: 0)</td>
</tr>
<tr>
<th>random_state</th>
<td>乱数のシードを固定する場合に指定。数値もしくは integer or numpy.RandomState で指定。 (デフォルト値: None)</td>
</tr>
<tr>
<th>copy_x</th>
<td>距離を事前に計算する場合、メモリ内でデータを複製してから実行するかどうか。 (デフォルト値: True)</td>
</tr>
<tr>
<th>n_jobs</th>
<td>初期化を並列処理する場合の多重度。-1 を指定するとすべての CPU を使用。 (デフォルト値: 1)</td>
</tr>
</table>

<p>scikit-learn を用いてクラスタ分析を行う手順を紹介します。</p>
<p><b>今回使用するデータ</b><br />
今回は、UC バークレー大学の <a href="http://archive.ics.uci.edu/ml/" target="_blank">UCI Machine Leaning Repository</a> にて公開されている、「<a href="https://archive.ics.uci.edu/ml/datasets/Wholesale+customers" target="_blank">Wholesale customers Data Set (卸売業者の顧客データ)</a>」を利用します。</p>
<p>データセットの構成は以下のようになっています。各行が顧客 1 件を指し、440 件の顧客データが格納されています。</p>
<table>
<tr>
<th>Channel</th>
<td>販売チャネル。1: Horeca (ホテル・レストラン・カフェ), 2: 個人向け小売</td>
</tr>
<tr>
<th>Region</th>
<td>各顧客の地域。1: リスボン市, 2: ポルト市, 3: その他</td>
</tr>
<tr>
<th>Fresh</th>
<td>生鮮品の年間注文額</td>

<th>Milk</th>
<td>生鮮品の年間注文額</td>

<tr>
<th>Grocery</th>
<td>食料雑貨の年間注文額</td>
</tr>
<tr>
<th>Frozen</th>
<td>冷凍食品の年間注文額</td>
</tr>
<tr>
<th>Detergents_Paper</th>
<td>衛生用品と紙類の年間注文額 </td>
</tr>
<tr>
<th>Delicassen</th>
<td>惣菜の年間注文額</td>
</tr>
</table>
<p>データセットの詳細は以下にて確認可能です。</p>
<ul>
<li><a href="https://archive.ics.uci.edu/ml/datasets/Wholesale+customers" target="_blank">UCI Machine Learning Repository: Wholesale customers Data Set</a></li>
<li><a href="http://pythondatascience.plavox.info/wp-content/uploads/2016/05/Wholesale_customers_data.csv">データセットのダウンロード (当サイトでミラー)</a></li>
</ul>
<p><b>分析用コード</b></p>
<p>以下のコードを実行して、クラスタ分析を実行できます。今回は、440 件の顧客を購買傾向に基づいて、4 つのクラスタに分類します。<br />
(※ K-Means 法は初期値に乱数を使用する関係上、必ずしも以下の結果通りにクラスタ番号が決定するとは限りません)</p>
<p></p><!-- Crayon Syntax Highlighter v_2.7.2_beta -->
</div>

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans

cust_df = pd.read_csv("Wholesale_customers_data.csv")
cust_df
```

```python
# 不要なカラムを削除
del(cust_df['Channel'])
del(cust_df['Region'])
```

```python
# Pandasのデータフレームから Numpy の行列 (Array) に変換
cust_array = np.array([cust_df['Fresh'].tolist(),
                       cust_df['Milk'].tolist(),
                       cust_df['Grocery'].tolist(),
                       cust_df['Frozen'].tolist(),
                       cust_df['Detergents_Paper'].tolist(),
                       cust_df['Delicassen'].tolist()
                      ], np.int32)

# 行列を転置
cust_array = cust_array.T

# クラスタ分析を実行(クラスタ数=4)
pred = KMeans(n_clusters=4).fit_predict(cust_array)
pred
```

<div style="font-size:20px; color:pink; margin:20px">

</div>
<div style="font-size:14px; color:white; margin:20px">
上記のように、440 件の各顧客にクラスタ番号 (0, 1, 2, 3) が付与されたことがわかります。
</div>

<div style="font-size:20px; color:pink; margin:20px">
各クラスタの特徴を確認
</div>
<div style="font-size:14px; color:white; margin:20px">
クラスタ分析の結果を利用し、各クラスタがどのような特徴があるのかを確認します。<br>
ここでは、集計作業を楽に行うため、Pandas のデータフレームを利用します。
</div>

```python
# Pandasのデータフレームワークにクラスタ番号を追加
cust_df['cluster_id']=pred
cust_df
```

```python
# 各クラスタに属するサンプル数の分布
cust_df['cluster_id'].value_counts()
```

```python
# 各クラスタの各部門商品の購買額の平均値(クラスタ番号=0)
cust_df[cust_df['cluster_id']==0].mean()
```

```python
# 各クラスタの各部門商品の購買額の平均値(クラスタ番号=1)
cust_df[cust_df['cluster_id']==1].mean()
```

```python
# 各クラスタの各部門商品の購買額の平均値(クラスタ番号=2)
cust_df[cust_df['cluster_id']==2].mean()
```

```python
# 各クラスタの各部門商品の購買額の平均値(クラスタ番号=3)
cust_df[cust_df['cluster_id']==3].mean()
```

<div style="font-size:20px; color:pink; margin:20px">
Matplotlib でクラスタの傾向を可視化
</div>
<div style="font-size:14px; color:white; margin:20px">
先ほど求めた、各クラスタの各部門商品の購買額の平均値をMatplotlibを用いて<br>
傾向を可視化すると以下のようになります。<br>
<br>
Matplotlib で積み上げ棒グラフを出力<br>
</div>

```python
# 可視化(積み上げ棒グラフ)
import matplotlib.pyplot as plt

clusterinfo = pd.DataFrame()
for i in range(4):
    clusterinfo['cluster' + str(i)] = cust_df[cust_df['cluster_id'] == i].mean()
clusterinfo = clusterinfo.drop('cluster_id')
clusterinfo
```

```python
my_plot = clusterinfo.T.plot(kind='bar', stacked=True, title="Mean Value of 4 Clusters")
my_plot.set_xticklabels(my_plot.xaxis.get_majorticklabels(), rotation=0)
```
